```python
import os

import comet_ml
from dotenv import load_dotenv

load_dotenv()
metric_keys = ["cdlib_modularity", "runtime", "customize_q0_overlap"]
api = comet_ml.API(api_key=os.getenv("COMET_API_KEY"))
experiments = api.get_experiments(
    workspace="d-ng-nguy-n",
    project_name="graph-community-detection-overlapping-bio-wormnet-v3",
)
metrics = []
runs = []
for exp in experiments:
    run = exp.get_parameters_summary()
    m = exp.get_metrics()
    filterd_m = [item for item in m if item["metricName"] in metric_keys]
    metrics.append(filterd_m)
    runs.append(run)
for metric in metrics:
    print(metric)
```
